```python
import torch 
import numpy as np
import pandas as pd
import this
```

```python
data=df.to_numpy()
print(data[0][0])
```

```python
X_txt=[]
Y=[]
for i in range(len(data)):
    data[i][0] = re.sub("<.*?>", " ", data[i][0])
    data[i][0] = re.sub("[^a-zA-Z]", " ", data[i][0])
    X_txt.append(data[i][0].lower().split())
    Y.append(data[i][1])
print(X_txt[0])
print(Y[0])
```

```python
all_tokens=[]
for i in X_txt:
    for j in i:
        all_tokens.append(j)

np.random.seed(42)
np.random.shuffle(all_tokens)

embed={}
counter =0
for i in all_tokens:
    if i not in embed:
        embed[i]=counter
        counter+=1

oh_data=[]
for i in X_txt:
    out=[]
    for j in i:
        out.append(embed[j])
    oh_data.append(out)
```

```python
max_len = 0
for i in X_txt:
    if len(i) > max_len:
        max_len = len(i)
print(max_len)
```

```python
from torch import tensor,long as ft
from torch.utils.data import Dataset,DataLoader
class QAdata(Dataset):
    def __init__(self,oh_data,ans,embed_l,max_len):
        self.oh=oh_data
        self.ans=ans
        self.e=embed_l
        self.m=max_len
        self.n=embed_l
    def __len__(self):
        return len(self.oh)
    def __getitem__(self,i):
        # oo=torch.ones(self.m, dtype=ft)*self.n
        # oo[:len(self.oh[i])] = torch.tensor(self.oh[i], dtype=ft)       
        # return oo,tensor(self.ans[i])
        return torch.tensor(self.oh[i], dtype=ft), tensor(self.ans[i], dtype=ft)
    
```

```python
X_train, y_train, X_test, y_test = oh_data[:40000], Y[:40000], oh_data[40000:], Y[40000:]
train_data = QAdata(X_train, y_train, embed_l=len(embed),max_len=max_len)
test_data = QAdata(X_test, y_test, embed_l=len(embed), max_len=max_len)
train_loader = DataLoader(train_data, batch_size=1, shuffle=True)
test_loader = DataLoader(test_data, batch_size=1, shuffle=False)
```

```python
import torch
i=0
for question, answer in train_data:
  print(question, answer)
  i+=1
  if i==10:
    break
```

```python
from torch import nn
class simpleLSTM(nn.Module):
    def __init__(self,vocab_size):
        super().__init__()
        self.embed=nn.Embedding(vocab_size+1,128)
        self.lstm=nn.LSTM(128,16,dropout=0.2,batch_first=True)
        self.lin=nn.Linear(16,1)
        self.sigmoid=nn.Sigmoid()
    def forward(self,X):
        em=self.embed(X)
        output,(hidden, state)=self.lstm(em)
        out=self.lin(output[:,-1,:])
        out=self.sigmoid(out)
        return out
```

```python
import time
from torch import optim
lr=0.001
epochs=20
model = simpleLSTM(len(embed))
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
model.to(device)
optimizer = optim.Adam(model.parameters(), lr=lr)
criterion = nn.BCEWithLogitsLoss()
for epoch in range(epochs):
    start_time = time.time()
    tl=0
    for X, y in train_loader:
        X = X.to(device)
        y = y.to(device)
        optimizer.zero_grad()
        output = model(X)
        loss = criterion(output.squeeze(0), y.float())
        loss.backward()
        optimizer.step()
        tl += loss.item()
    elapsed = time.time() - start_time
    print(f"Epoch {epoch+1}/{epochs}, Loss: {tl/len(train_loader)}, Time: {elapsed:.2f}s")
```

```python
i=0
for X,y in test_loader:
    output = model(X.to(device))
    output = output.squeeze(0).cpu().detach().numpy()
    str=""
    print("X:",X.shape)
    for i in X[0]:
        str += re[i.item()] + " "
    print(f"Input: {str.strip()}, Predicted: {output}, Actual: {y.item()}")
    i+=1
    if i==20:
        break
    
```

```python
tp=0
fp=0
fn=0
tn=0
for X,y in test_loader:
    output = model(X.to(device))
    output = output.squeeze(0).cpu().detach().numpy()
    if output >= 0.5:
        if y.item() == 1:
            tp += 1
        else:
            fp += 1
    else:
        if y.item() == 0:
            tn += 1
        else:
            fn += 1
```

```python
print(f"True Positives: {tp}, False Positives: {fp}, \n True Negatives: {tn}, False Negatives: {fn}")
```

```python
accuracy = (tp + tn) / (tp + fp + tn + fn)
print(f"Test Accuracy: {accuracy:.4f}")
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
print(f"Test Precision: {precision:.4f}")
```
